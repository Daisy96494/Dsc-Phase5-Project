```python
import pandas as pd
```

```python
data = pd.read_csv('weatherhistoricaldata1.csv')
```

```python
data.columns
data['WND']
```

```python
def parseTemperature(entry):
    temp_str, _ = entry.split(",")
    return int(temp_str)/10

```

```python
def parseRainfall(entry):
    parts = entry.split(",")
    if len(parts) != 4:
        return None
    precipitation = int(parts[1])/10

    return precipitation
    
```

```python
def parseCeilingHeight(entry):
    parts = entry.split(",")
    if len(parts) != 4:
        return None
    
    celheight = None if parts[0] == "99999" else int(parts[0])

    return celheight
```

```python
def parseVisibility(entry):
    parts = entry.split(",")
    if len(parts) != 4:
        return None
    
    vis = None if parts[0] == "99999" else int(parts[0])/10 ##review

    return vis
```

```python
def parseSLPressure(entry):
    parts = entry.split(",")
    if len(parts) != 2:
        return None
    
    slp = None if parts[0] == "99999" else int(parts[0])/10

    return slp
```

```python
def parseDew(entry):
    parts = entry.split(",")
    if len(parts) != 2:
        return None
    
    dew = int(parts[0])

    return dew
```

```python
def parseWindSpeed(entry):
    parts = entry.split(",")
    if len(parts) != 5:
        return None
    
    wind = None if parts[3] == "99999" else int(parts[3])/10  ## wind in m/s

    return wind
```

```python
def parseWindDir(entry):
    parts = entry.split(",")
    if len(parts) != 5:
        return None
    
    direction = None if parts[0] == "99999" else int(parts[0])

    return direction
```

```python
## Parsing Block

data['WND'] = data['WND'].apply(parseWindSpeed)
# data['CIG'] = data['CIG'].apply(parseCeilingHeight)
data['AA1'] = data['AA1'].apply(parseRainfall)
data['TMP'] = data['TMP'].apply(parseTemperature)
data['SLP'] = data['SLP'].apply(parseSLPressure)
data['DEW'] = data['DEW'].apply(parseDew)
```
